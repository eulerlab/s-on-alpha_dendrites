```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import os
```

# Data

```python
from djimaging.user.alpha.utils.populate_alpha import PROJECT_ROOT
```

```python
filepath_calcium = os.path.join(PROJECT_ROOT, 'data/Ran/convex_hull_calcium.csv')
df_calcium = pd.read_csv(filepath_calcium)
df_calcium['indicator'] = 'Calcium'
df_calcium
```

```python
filepath_glutamate = os.path.join(PROJECT_ROOT, 'data/Ran/convex_hull_glutamate.csv')
df_glutamate = pd.read_csv(filepath_glutamate)
df_glutamate['indicator'] = 'Glutamate'
df_glutamate
```

```python
df = pd.concat([df_calcium, df_glutamate], ignore_index=True)
```

# Plot

```python
from alphaanalysis import plot as plota

plota.set_default_params(kind='paper')
```

```python
FIGURE_FOLDER = './figures'
```

```python
palette = plota.get_palette('calcium')
order = plota.get_order('calcium')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(3, 1.5))

sns.despine(bottom=True)
sns.boxplot(ax=ax, data=df, y='hull_cdia_um', x='group', palette=palette, order=order, whis=False, fliersize=0)
sns.swarmplot(ax=ax, data=df, y='hull_cdia_um', x='group', color='k', order=order, hue='indicator', palette='gray',
              alpha=0.8, size=5)

ax.set(ylabel='Dendritic field\ndiameter [Âµm]', xlabel=None)
ax.tick_params(axis='x', length=0)
ax.legend(bbox_to_anchor=(1, 1), loc='upper left')

plt.tight_layout()

plt.savefig(os.path.join(FIGURE_FOLDER, 'dend_sizes.pdf'))
```
